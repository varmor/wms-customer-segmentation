```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
customer_data = pd.read_csv("../datasets/final-wms-customer-dataset.csv")
customer_data
```

```python
features = ["sip_amount", "networth", "age"]
```

```python
data = customer_data[features]
data.head()
```

```python
# scaling the data
data = ((data - data.min())/(data.max() - data.min())) * 9 + 1
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(3)
kmeans.fit(data)
```

```python
centroids2 = kmeans.cluster_centers_
labels = kmeans.predict(data)
```

```python
labels
```

```python
centroids2
```

```python
pd.DataFrame(centroids2, columns=features).T
```

```python
label_0 = customer_data[labels == 2][["sip_amount","networth","age" ]]
label_0
```
